```python
from time import sleep
import yaml
from geopy.geocoders import Nominatim


with open('../src/assets/courts.yaml', 'r', encoding='utf-8') as file:
    tennis_clubs = yaml.safe_load(file)

# Initialize geolocator
geolocator = Nominatim(user_agent="tennis_club_locator")

# Find coordinates for each club
for club in tennis_clubs:
    location = geolocator.geocode(club["address"], exactly_one=True)
    sleep(1)
    if location:
        club["coordinates"] = {
            "lat": location.latitude,
            "lng": location.longitude
        }
    else:
        print(f"Coordinates could not be found for address: {club['address']}")
```

```python
print("Clubs without coordinates")
clubs_without_coordinates = [club for club in tennis_clubs if "coordinates" not in club]
print([club["name"] for club in clubs_without_coordinates])

print("Yaml:")
print(yaml.dump(tennis_clubs, allow_unicode=True, default_flow_style=False, sort_keys=False))
```
